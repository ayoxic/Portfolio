Ecrire un programme qui affiche tous les éléments dupliqués d'une liste données sans utiliser aucune fonction prédéfinie en Python.<br>

    * indication : utiliser une double boucle for

```python
# Exemple
# L = [7 , 21 , 5 , 12 , 7 , 91 , 21 ,  12 , 28]
# la sortie est: [7, 21, 12]
L = [7 , 21 , 5 , 12 , 7 , 91 , 21 ,  12 , 28]
doublons = []
for i in range(len(L)):
    for j in range(i + 1, len(L)):
        if L[i] == L[j]:
            deja_vu = False
            for x in doublons:
                if x == L[i]: deja_vu = True
            if not deja_vu:
                doublons.append(L[i])
print(doublons)
```

- Écrire un programme qui prend en paramètre une liste d’entiers T et  et qui affiche la moyenne des éléments de T.


```python
import random
T = [random.randint(0, 10) for i in range(5)]
print(T)
```

```python
# solution 1 : En utilisant une boucle
T = [10, 20, 30, 40, 50]
somme = 0
compteur = 0
for nombre in T:
    somme = somme + nombre
    compteur = compteur + 1
if compteur != 0:
    moyenne = somme / compteur
    print("Moyenne :", moyenne)
else:
    print("La liste est vide")
```

```python
# solution 2 : En utilisant  sum()

T = [10, 20, 30, 40, 50]
total = sum(T)
nb_elements = 0
for _ in T:
    nb_elements += 1
if nb_elements > 0:
    print("Moyenne :", total / nb_elements)
```

- Scinder la liste T en deux listes, une contenant les éléments supérieurs à la moyenne et l'autre contenant les éléments inférieure à la moyenne

```python
T = [10, 20, 30, 40, 50, 5, 45]
moyenne = sum(T) / len(T)
print("La moyenne est :", moyenne)
T_superieur = []
T_inferieur = []
for x in T:
    if x > moyenne:
        T_superieur.append(x)
    elif x < moyenne:
        T_inferieur.append(x)
print("Éléments supérieurs à la moyenne :", T_superieur)
print("Éléments inférieurs à la moyenne :", T_inferieur)
```

<b>Exemple</b>:
L = [5, 6, 9, 8, 10]
L devient :
L = [6, 8, 10, 5, 9] ou L = [8, 6, 10, 9, 5]

<b>N.B.</b>
    - l'ordre interne des paires ou impaires n'est pas interessant
    - Ne pas utiliser d'espaces auxiliaires (l'arrangement doit se faire <b><u>inPlace</u></b>)
</pre>


```python
import random
L = [random.randint(0, 20) for i in range(10)]
L
```

```python
# ecrire votre solution ici

L = [5, 6, 9, 8, 10]
gauche = 0
droite = len(L) - 1
while gauche < droite:
    if L[gauche] % 2 == 0:
        gauche += 1
    elif L[droite] % 2 != 0:
        droite -= 1
    else:
        L[gauche], L[droite] = L[droite], L[gauche]
        gauche += 1
        droite -= 1
print("L après réarrangement :", L)
    
```

- Écrire un programme qui remplace les slashes dans une chaine par des backslashes.
<br>Exemple : 
    - chemin = "D:/data/info/prog/python"    devient
    - chemin = "D:\data\info\prog\python"

```python
ch = "D:/data/info/prog/python"
print(ch)
```

```python
# ecrire votre solution ici
chemin = "D:/data/info/prog/python"
nouveau_chemin = ""
for caractere in chemin:
    if caractere == "/":
        nouveau_chemin += "\\"
    else:
        nouveau_chemin += caractere
print("Ancien :", chemin)
print("Nouveau :", nouveau_chemin)
    
```
